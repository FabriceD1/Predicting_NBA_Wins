###PER GAME STATS

#Create List of the years
Lets do 10 years of nba data from 2013 to 2023

Set the range from 2013-2024

```python
years = list(range(2013, 2024))
```

Lets see our list by calling the variable years

```python
years
```

###Lets create a vairable for the url we want to scrape

```python
url_start = "https://www.basketball-reference.com/leagues/NBA_{}.html"
```

Create a folder to store our html code for each website into our working directory

```python
!mkdir nba_stats
```

Send a request to obtain the html code

```python
import requests

for year in years:
  url = url_start.format(year)
  data = requests.get(url)

  with open("nba_stats/{}.html".format(year), "w+") as f:
    f.write(data.text)
```

```python
from bs4 import BeautifulSoup
```

```python
with open("nba_stats/2013.html") as f:
  page = f.read()
```

```python
soup = BeautifulSoup(page, "html.parser")
```

If we look at the Per Game Stat table on the BasketballReference Page, we notice there is an extra row to seperate the bottom half of the teams. We do not want an extra header row when upload our data to PANDAS. Thus, we will use soup find to look for the specific tr code that has a class called thead. Decompose will remove any elements that we find

```python
soup.find('tr', class_= "thead").decompose()
```

```python
soup.find('tfoot').decompose()
```

Because we only want the Per Game Stat table, lets remove all other elements from the page that we will not need

```python
stat_table = soup.find(id="per_game-team")
```

```python
import pandas as pd
```

```python
stat_2013 = pd.read_html(str(stat_table))[0]
```

```python
stat_2013
```

```python
dfs = []

for year in years:
  with open("nba_stats/{}.html".format(year)) as f:
    page = f.read()
  soup = BeautifulSoup(page, "html.parser")
  soup.find('tr', class_="thead").decompose()
  soup.find('tfoot').decompose()
  stat_table = soup.find(id="per_game-team")
  stat_years = pd.read_html(str(stat_table))[0]
  stat_years["Year"] = year

  dfs.append(stat_years)
```

```python
dfs
```

Lets combine all of the list of dataframes and combine them into one

```python
df_stats = pd.concat(dfs)
df_stats
```

```python
df_stats.to_csv("nba_stats.csv")
```

###ADVANCED STAT STATISITICS

```python
dfs_advanced = []

for year in years:
  with open("nba_stats/{}.html".format(year)) as f:
    page = f.read()
  soup = BeautifulSoup(page, "html.parser")
  soup.find('tr', class_="over_header").decompose()
  soup.find('tfoot').decompose()
  advanced_table = soup.find(id="advanced-team")
  advanced_years = pd.read_html(str(advanced_table))[0]
  advanced_years["Year"] = year

  dfs_advanced.append(advanced_years)
```

```python
dfs_advanced
```

```python
advanced_stats = pd.concat(dfs_advanced)
advanced_stats
```

```python
advanced_stats = advanced_stats.drop(columns=['Unnamed: 17','Unnamed: 22','Unnamed: 27'])
advanced_stats_cleaned = advanced_stats.loc[advanced_stats['Team'] != 'League Average']
```

```python
advanced_stats_cleaned
```

```python
advanced_stats_cleaned.to_csv("advanced_stats.csv")
```

Merge the two data sets by Team and the Year

```python
# merge the two data sets by team name and year
merged_data = pd.merge(df_stats, advanced_stats, on=["Team", "Year"])
```

# Remove the Unnamed, Rk, Arena, Attend., and Attend./G columns

```python
merged_data = merged_data.loc[:, ~merged_data.columns.str.contains('^Unnamed|^Rk|^Arena$|^Attend.$|^Attend./G$|^G$')]
```

```python
merged_data
```

# Replace the asterisk character in team names

```python
merged_data['Team'] = merged_data['Team'].str.replace('*', '')
```

```python
merged_data
```

```python
# save the merged data as a new csv file
merged_data.to_csv("merged_data.csv", index=False)
```
